```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
headers = ['SubjectArea','F13','M13','other13', '2013/14', 'F14','M14','other14','2014/15','F15','M15','other15', '2015/16','F16','M16','other16', '2016/17','F17','M17','other17', '2017/18']
df = pd.read_csv('test.csv', names = headers)
df = df.set_index('SubjectArea')


# Number of students per subject area between 2013-2018
df.plot.bar( y = ['2013/14','2014/15','2015/16','2016/17','2017/18'],figsize=(20,5))
plt.title('Number of students per subject area between 2013-2018')
plt.show()




# Number of Female students per subject area between 2013-2018
df.plot.bar( y = ['F13','F14','F15','F16','F17'],figsize=(20,5))
plt.title('Number of Female students per subject area between 2013-2018')
plt.show()


# Number of Male students per subject area between 2013-2018
df.plot.bar( y = ['M13','M14','M15','M16','M17'],figsize=(20,5))
plt.title('Number of Male students per subject area between 2013-2018')
plt.show()

```
